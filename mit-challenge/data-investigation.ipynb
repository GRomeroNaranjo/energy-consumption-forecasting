## Importing the libraries and reading the data

```python
import torch
from torch import nn
import torch.optim as optim
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

data = pd.read_csv("PJM(AEP)_energy_dataset.csv")
data.head()
```

## Describing the data with some mathematical calculations

```python
data.describe()
```

## Plotting the hourly energy consumptions

```python
data = data.sort_values(by="Datetime")

plt.figure(figsize=(20, 6))
plt.plot(data["PJME_MW"])
plt.xlabel("Hour")
plt.ylabel("Energy Consumption MW")
plt.title("Hourly energy Consumption PJM")
plt.show()
```

## Plotting the hourly yearly consumptions

```python
data = data.sort_values(by="Datetime")

data["year"] = data["Datetime"].str[:4]
unique_years = data.drop_duplicates(subset="year", keep="first")

years = unique_years["year"]
value = unique_years["PJME_MW"]

dataframe = pd.DataFrame({
    "date": years,
    "value": value,
})

dataframe = dataframe.reset_index(drop=True)

x = dataframe["date"]
y = dataframe["value"]

plt.figure(figsize=(12, 6))
plt.plot(x, y)
plt.xlabel("Year")
plt.ylabel("Energy Consumption MW")
plt.title("Yearly energy Consumption PJM")
plt.show()
```
